```python
# Import necessary libraries
import fitz
import pandas as pd
```

```python
# Use PyMuPDF to read the PDF file.
doc = fitz.open("./data/order.pdf")
page = doc[0]
```

```python
# Find the tables in the PDF and print the column names.
tabs = page.find_tables()
for i,tab in enumerate(tabs):
    print(f"Table {i} column names: {tab.header.names}")
```

```python
# Convert the first table to a Pandas DataFrame and print it.
tab = tabs[0]
df = tab.to_pandas()
df
```

```python
# Clean up the DataFrame by removing newlines.
df = df.replace('\n', ' ', regex=True)

# Reformat the DataFrame to be more readable by language models and save it to a text file.
with open("./processed_data/table_data.txt", "w") as f:
    for index, row in df.iterrows():
        row_str = ""
        for col in df.columns:
            row_str += f"{col}: {row[col]}, "
        formatted = row_str[:-2]
        print(formatted)
        f.write(formatted + "\n")
```
